```python
import os
import glob

file_dir = '../../出租车数据共享_2019CUMCM/taxi_data'

# 使用glob读取所有的csv文件

file_list = glob.glob(os.path.join(file_dir, '*.csv'))

print(len(file_list))
```

```python
import pandas as pd

# 使用pandas读取依次读取csv文件 检查time列的前四位是不是2011
for file in file_list:
    df = pd.read_csv(file, header=0, encoding='gbk')

    year_lists = df.time.str[:4].unique().tolist()

    if '2011' not in year_lists:
        print('2011 not in year_lists')
        print(file)
        break
    else:
        print('2011 in year_lists')
        print(file)
```

深圳市出租车轨迹数据说明:

1.数据采集时间段为2011/04/18－2011/04/26,车辆总数为13798台。
注：因行车环境的影响，记录采集间隔时间周期不恒定。

2.每个车辆轨迹数据文件均以车牌号命名，数据文件采用csv格式存储，数据文件由7个字段构成，分别如下：

name 车牌号
time 采集时间点(格式：YYYY/MM/DD hh:mm:ss)
jd 经度
wd 纬度
status 车辆状态（0＝非打表，即：空载；1＝已打表，即：重载）
v 车速（单位为：km/h)
angle 行车方向（0＝东；1＝东南；2＝南；3＝西南；4＝西；5＝西北；6＝北；7＝东北）
